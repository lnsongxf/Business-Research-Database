```python
import pandas as pd
```

```python
df=pd.read_csv('in.csv')
df['source']=df['source'].str.replace(' ','_')
df['target']=df['target'].str.replace(' ','_')
df['weight']=df['weight']*100
df
```

```python
dg=df.pivot(index='source',columns='target')
dg=dg.fillna(0)
dg
```

```python
dg.to_csv('out.csv',sep='\t',float_format='%.0f' )
```
