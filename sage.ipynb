```python
import urllib2
import requests
import codecs
import os
import random
import time
from bs4 import BeautifulSoup as bs
import re
```

```python
base_folder='src/sae/'+['jrnmng/','orgstd/','adscqt/','humrel/'][3]
jrnl=['?/','?/','asqa/','huma/'][3]

base_url='http://journals.ama.org/'
base_vol_iss=base_url+'loi/'+jrnl
base_htmls=base_url+'toc/'+jrnl
```

```python
#get vol_issues
session = requests.Session()
response = session.get(base_vol_iss)
html=response.text
s=bs(html,'html.parser')
vol_iss_links=[l['href'] for l in s.findAll('a',{"href":re.compile(base_htmls)})]
vol_iss_links=list(set(vol_iss_links))
len(vol_iss_links)
```

```python
#get htmls from saga
vol_iss_folder=base_folder+'vol_iss/'
for link in vol_iss_links:
    v=link.strip(base_htmls)
    print v,
    page=str(v)
    file_name=vol_iss_folder+page.replace('/','_')+'.html'
    if not os.path.isfile(file_name):
        url=base_htmls+page
        response = session.get(url)
        html=response.text
        f = codecs.open(file_name, 'w','utf-8')
        f.write(html) 
        f.close()        
```

```python
dois=[]
for f in os.listdir(vol_iss_folder):
    if 'html' in f:
        print f, 
        html=open(vol_iss_folder+f)
        s = bs(html, 'html.parser')
        links=s.find_all('div',{'class':'art_title linkable'})
        dois=dois+[l.findAll('a')[0]['href'].replace('/doi/full/','').replace('/doi/pdf/','') for l in links] 
dois=list(set(dois))
len(dois)
```

```python
base='http://journals.sagepub.com/doi/abs/'
folder=base_folder+'htmls/'
session = requests.Session()
i=0
for doi in dois:
    print i,
    i=i+1
    title=doi.replace('/','_')+'.html'
    if not os.path.isfile(folder+title):
        url= base+doi
        response = session.get(url)
        html=response.text
        f = codecs.open(folder+title, "w", "utf-8")
        f.write(html,) 
        f.close()
        #r=random.uniform(1, 3)
        #time.sleep(r) # delays for 3 to 4 seconds
```
